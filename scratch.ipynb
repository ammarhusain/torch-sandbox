```python
from torch import nn
import torch
loss_fn = nn.BCELoss()
real_tgt = torch.ones(10, )

lin = nn.Linear(100, 10)
sig = nn.Sigmoid()
data = torch.zeros(100, )

e = loss_fn(sig(lin(data)), real_tgt)

e.backward

print(e)
print(lin.weight.grad)
```
